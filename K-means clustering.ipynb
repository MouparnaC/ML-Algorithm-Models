K-means Model Building

Elbow Method

```python
from sklearn.cluster import KMeans
from sklearn import datasets, metrics, cluster
from sklearn.metrics import silhouette_score

Centroid_list=[]
no_of_clusters=range(1,21)

for i in no_of_clusters:
  kmeans=cluster.KMeans(n_clusters=i, init="k-means++", max_iter=300, n_init=10, random_state=0)
  kmeans.fit(df.values)
  Centroid_list.append(kmeans.inertia_)


plt.plot(no_of_clusters,Centroid_list)
plt.title("Elbow Method")
plt.show()
```

Here we are plotting the Elbow Method. We can see the kink/elbow part approximately at point 2. So the optimum cluster would be 2.

Training and Predictions

```python
kmeans=cluster.KMeans(n_clusters=2, init="k-means++", max_iter=300, n_init=10, random_state=0)
Centroid= kmeans.fit_predict(df.values)
```

```python
#print the centroid values
print(Centroid)
```

Plotting the predictions

```python
plt.figure(figsize=(12,10))
plt.scatter(df.values[Centroid== 0,0],df.values[Centroid== 0,1], s=100, c="red", label="Malignant")
plt.scatter(df.values[Centroid== 1,0],df.values[Centroid== 1,1], s=100, c="green", label="Beningn")

plt.scatter(kmeans.cluster_centers_[1,0],kmeans.cluster_centers_[1,1], s=100, c="yellow", label="Centroids")
plt.legend()
```

Silhouette Score

```python
print(silhouette_score(df,kmeans.labels_)) 
```

So Silhouette Score is 69%
